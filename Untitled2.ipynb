```python
import numpy as np

A = np.array([
    [1, 1, 1, 1],
    [2, -1, 1, 3],
    [-1, 2, -1, 1],
    [3, 1, 2, -1]
])

B = np.array([10, 20, -5, 15])

# Vérifier le déterminant
det = np.linalg.det(A)
print("Determinant:", det)

# Résolution
X = np.linalg.solve(A, B)
print("Solution [x, y, z, w]:", X)
```
