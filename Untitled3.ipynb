```python
import sympy as sp

x = sp.symbols('x')      # dclare le symbole x
f = x**3 - 6*x**2 + 9*x + 1   # dfinition de la fonction

# Drive
f_prime = sp.diff(f, x)

print("f(x) =", f)
print("f'(x) =", f_prime)
```

```python
crit_points = sp.solve(f_prime, x)
print("Points critiques :", crit_points)
```

```python
racines = sp.solve(f, x)
print("Racines de f(x) =", racines)
```

```python
lim_inf = sp.limit(f, x, -sp.oo)
lim_sup = sp.limit(f, x, sp.oo)

print("lim x- =", lim_inf)
print("lim x+ =", lim_sup)
```

```python
from sympy.calculus.singularities import singularities
from sympy.calculus.util import continuous_domain, function_range

# Table de variation approximative
table_variation = function_range(f, x, sp.Interval(-sp.oo, sp.oo))
print("Image / variations :", table_variation)
```

```python
sign_fprime = sp.solve(sp.Eq(f_prime,0), x)
print("Signe de f'(x) s'annule en :", sign_fprime)
```

```python
import numpy as np
from scipy import linalg, stats

A = np.array([[2, 3], [1, 4]])
b = np.array([5, 6])

# Dterminant
print("det(A) =", linalg.det(A))

# Inverse
print("A^-1 =", linalg.inv(A))

# Rsolution systme
x = linalg.solve(A, b)
print("Solution x =", x)

# Valeurs propres
vals, vecs = linalg.eig(A)
print("Valeurs propres =", vals)

# Moyenne et std
print("Moyenne =", stats.tmean(A))
print("Ecart-type =", stats.tstd(A))
```

```python
import numpy as np
import pandas as pd
from scipy import stats

data = np.random.normal(10, 2, size=1000)
df = pd.DataFrame(data, columns=["X"])

# Statistiques de base
print("Moyenne :", df["X"].mean())
print("Mdiane :", df["X"].median())
print("cart-type :", df["X"].std())
print("Variance :", df["X"].var())
print("Skewness :", stats.skew(df["X"]))
print("Kurtosis :", stats.kurtosis(df["X"]))
print("Quantiles :", df["X"].quantile([0.25,0.5,0.75]))
```
